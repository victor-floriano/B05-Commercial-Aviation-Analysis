#### **What is the average departure and arrival delay for every airline?**

---

While analyzing the average departure and arrival delays for each airline, we found distinct patterns of punctuality, showing a broad picture in terms of operational efficiency.

Spirit Air Lines and United Air Lines showcase longer average departure delays at 15.94 and 14.44 minutes, respectively. In contrast, Hawaiian Airlines and Alaska Airlines boast excellent punctuality, with average departure delays as low as 0.49 and 1.79 minutes. Notably, Delta Air Lines stands out for its minimal average arrival delay of 0.19 minutes, suggesting a remarkable ability to maintain on-time arrivals. In contrast, Frontier Airlines and JetBlue Airways experience more substantial average arrival delays at 12.50 and 6.68 minutes, respectively. Lastly, Alaska Airlines has a negative average arrival delay, meaning its flights land about 1 minute earlier than the scheduled time.

These differences underscore the operational divergences across airlines, as some demonstrate exceptional proficiency in mitigating delays, while others struggle with more visible deviations from their scheduled timelines.

This initial question of the project not only allows us to visualize the diverse landscape of airlines in terms of punctuality but also serves as a crucial lens through which we can discover insights into their operational effectiveness and commitment to timely performance.

```python
# Calculating average departure and arrival delays by airline

%%bigquery --project = ba775-b05

SELECT
  B.Airline,
  ROUND(AVG(A.DEPARTURE_DELAY), 2) AS avg_departure_delay,
  ROUND(AVG(A.ARRIVAL_DELAY), 2) AS avg_arrival_delay
FROM
  ba775-b05.airlines_kaggle.flights A
  JOIN ba775-b05.airlines_kaggle.airlines B
  ON A.AIRLINE = B.IATA_Code
GROUP BY
  B.Airline
ORDER BY
  avg_departure_delay DESC, avg_arrival_delay DESC;
```

#### **What is the distribution of flight delay times?**

---

Our analysis of flight delay times reveals a diverse distribution of delays, ranging from early arrivals to significant delays. The majority of flights (3,497,001 instances) actually arrive early, indicating operational efficiency. The next largest categories include delays of over an hour (616,659 instances), followed by shorter delays ranging from 5 to 15 minutes (438,705 instances), 15 to 30 minutes (369,068 instances), and 30 to 60 minutes (367,420 instances). Notably, delays of less than 5 minutes are also common (342,230 instances), and a small portion of flights (82,925 instances) experience no delay at all. This detailed breakdown provides a comprehensive view of delay durations, highlighting the problem that the most common delaye category is the longer `1h+ delay`, which could cause significant operational problems forr both the airlines and the airports.

Understanding the distribution of flight delay times is also useful for refining our project's analysis. From a business perspective a detailed categorization also enables better resource allocation, since strategies can differ for managing short versus long delays.

```python
# Most flights are actually early, and the most common delay time is 1h+ delays - we could change the break points to check for different things

%%bigquery --project = ba775-b05

SELECT
  CASE
    WHEN departure_delay IS NULL OR arrival_delay IS NULL THEN 'missing data'
    WHEN (departure_delay + arrival_delay) > 60 THEN '1h+ delay'
    WHEN (departure_delay + arrival_delay) > 30 THEN '30 to 60 min delay'
    WHEN (departure_delay + arrival_delay) > 15 THEN '15 to 30 min delay'
    WHEN (departure_delay + arrival_delay) > 5 THEN '5 to 15 min delay'
    WHEN (departure_delay + arrival_delay) > 0 THEN 'less than 5min delay'
    WHEN (departure_delay + arrival_delay) = 0 THEN 'no delay'
    WHEN (departure_delay + arrival_delay) < 0 THEN 'early'
    END AS delay,
  COUNT(*) delay_count
FROM
  ba775-b05.airlines_kaggle.flights
GROUP BY
  delay
ORDER BY
  delay_count DESC;
```

#### **Distribution of Delays by Airline**

---

To answer this question we chose to retrieve the percentage of delays that falls into each delay category by airline. We used the following categories:

* '1h+ delay'
* '30 to 60 min delay'
* '15 to 30 min delay'
* '5 to 15 min delay'
* 'less than 5min delay'

This analysis, which examines the distribution of delay types across different airlines, highlights the unique challenges each airline faces in managing flight delays. For example, a comparison between American Airlines and JetBlue reveals distinct patterns in their delay types. While 28.9% of American Airlines' delays are classified as severe, this figure rises to 34.4% for JetBlue. Additionally, while severe delays represent the highest percenatge of delays for American Airlines and JetBlue, for Alaska Airlines, the most common delay type is 5 to 15 minutes, a much shorter delay.

From a commercial perspective, this suggests that airlines would greatly benefit from a deeper, more personalized analysis of their specific delay issues. Such an approach would enable each airline to strategically allocate resources and develop targeted solutions to efficiently address the various types of delays they encounter, be they severe or short delays.

Note: Given the limiting way that pandas DataFrames display data in Colab ennvironments, we have provided a screenshot of the query results from BigQuery for easier reference.

```python
#Shows the distribution of delays by airline. Here early, on-time and null were
#excluded to show a clearer visualization

%%bigquery --project = ba775-b05

WITH dist_delays AS (
SELECT
  airline,
  delay_category,
  ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (PARTITION BY airline), 1) AS delay_percentage
FROM (
  SELECT
    airline,
    CASE
      WHEN departure_delay IS NULL OR arrival_delay IS NULL THEN 'missing data'
      WHEN (departure_delay + arrival_delay) > 60 THEN '1h+ delay'
      WHEN (departure_delay + arrival_delay) > 30 THEN '30 to 60 min delay'
      WHEN (departure_delay + arrival_delay) > 15 THEN '15 to 30 min delay'
      WHEN (departure_delay + arrival_delay) > 5 THEN '5 to 15 min delay'
      WHEN (departure_delay + arrival_delay) > 0 THEN 'less than 5min delay'
      WHEN (departure_delay + arrival_delay) = 0 THEN 'no delay'
      WHEN (departure_delay + arrival_delay) < 0 THEN 'early'
    END AS delay_category
  FROM
    ba775-b05.airlines_kaggle.flights
) AS delay_data
GROUP BY
  airline, delay_category
HAVING delay_category IN ('1h+ delay','30 to 60 min delay','15 to 30 min delay','5 to 15 min delay','less than 5min delay')
ORDER BY
  airline, delay_percentage DESC
)


SELECT
    airline,
    ARRAY_AGG(STRUCT(delay_category, delay_percentage)) AS aggregated_data
FROM
    dist_delays
GROUP BY
    airline;
```

```python
# For the top 5 airports with the most delays (departure/arrival), what are the top 5 airlines with the most delays?

%%bigquery --project = ba775-b05

# Departure delays

SELECT
  B.ORIGIN_AIRPORT,
  AL.Airline,
  B.delay_by_airport AS AIRPORT_DELAYS,
  B.delay_by_airport_airline AS AIRLINE_DELAYS
FROM
(
  SELECT A.*,
  RANK() OVER(PARTITION BY ORIGIN_AIRPORT ORDER BY delay_by_airport_airline DESC) most_delays_rank
  FROM
  (
    SELECT DISTINCT
    S.ORIGIN_AIRPORT,
    S.AIRLINE,
    SUM(delay) OVER(PARTITION BY ORIGIN_AIRPORT) delay_by_airport,
    SUM(delay) OVER(PARTITION BY ORIGIN_AIRPORT, AIRLINE) delay_by_airport_airline
    FROM
      (
      SELECT *,
      IF(DEPARTURE_DELAY > 0, 1, 0) AS delay
      FROM ba775-b05.airlines_dataset.flights
      ) AS S
    WHERE S.delay = 1
  ) A
) B
  JOIN ba775-b05.airlines_dataset.airlines AL
  ON B.AIRLINE = AL.IATA_Code
WHERE
  B.most_delays_rank <= 3
ORDER BY
  B.delay_by_airport DESC, B.delay_by_airport_airline DESC
LIMIT 15;
```

```python
# Arrival delays

%%bigquery --project = ba775-b05

SELECT
  B.DESTINATION_AIRPORT,
  AL.Airline,
  B.delay_by_airport AS AIRPORT_DELAYS,
  B.delay_by_airport_airline AS AIRLINE_DELAYS
FROM
(
  SELECT A.*,
  RANK() OVER(PARTITION BY DESTINATION_AIRPORT ORDER BY delay_by_airport_airline DESC) most_delays_rank
  FROM
  (
    SELECT DISTINCT
    S.DESTINATION_AIRPORT,
    S.AIRLINE,
    SUM(delay) OVER(PARTITION BY DESTINATION_AIRPORT) delay_by_airport,
    SUM(delay) OVER(PARTITION BY DESTINATION_AIRPORT, AIRLINE) delay_by_airport_airline
    FROM
      (
      SELECT *,
      IF(ARRIVAL_DELAY > 0, 1, 0) AS delay
      FROM ba775-b05.airlines_dataset.flights
      ) AS S
    WHERE S.delay = 1
  ) A
) B
  JOIN ba775-b05.airlines_dataset.airlines AL
  ON B.AIRLINE = AL.IATA_Code
  WHERE B.most_delays_rank <= 3
ORDER BY
  B.delay_by_airport DESC, B.delay_by_airport_airline DESC
LIMIT 15;
```

#### **What are the months with highest number of delayed flights (cumulative and proportionally)?**

---


In our analysis of the flight delays dataset, we identified that July, August, and March have the highest number of cumulative flight delays, while June, February, and July have the highest percentage of delayed flights, around 40%. These insights are particularly valuable for resource allocation and management in airports and airlines. During these peak delay months, additional resources can be allocated to areas most affected by delays, such as customer service, ground operations, and maintenance. By anticipating the increased workload, airports and airlines can better manage passenger flow, reduce congestion, and minimize the cascading effects of delays.

Additionally, schedule adjustments based on these findings can significantly mitigate the impact of delays. Airlines might consider revising flight schedules to include longer layovers or buffer times between flights during these high-risk months.

**Note:** For this analysis we considered a flight to be delayed it the sum of its departure delay and arrival delay was larger than 1 (min).

```python
# Months with highest number of delayed flights

%%bigquery --project = ba775-b05

WITH n_delays_per_month AS
(
  SELECT
    month,
    SUM(CASE WHEN (departure_delay + arrival_delay) > 0 THEN 1
        ELSE 0 END) AS n_flights_delayed,
  FROM
    ba775-b05.airlines_kaggle.flights
  GROUP BY
    month
),

total_flights_per_month AS
(
  SELECT
    month,
    COUNT(*) AS total_flights
  FROM
    ba775-b05.airlines_kaggle.flights
  GROUP BY
    month
)

SELECT
  CASE
    WHEN n_delays_per_month.month = 1 THEN 'Jan' -- Changes the month values to a more readable format
    WHEN n_delays_per_month.month = 2 THEN 'Feb'
    WHEN n_delays_per_month.month = 3 THEN 'Mar'
    WHEN n_delays_per_month.month = 4 THEN 'Apr'
    WHEN n_delays_per_month.month = 5 THEN 'May'
    WHEN n_delays_per_month.month = 6 THEN 'Jun'
    WHEN n_delays_per_month.month = 7 THEN 'Jul'
    WHEN n_delays_per_month.month = 8 THEN 'Aug'
    WHEN n_delays_per_month.month = 9 THEN 'Sep'
    WHEN n_delays_per_month.month = 10 THEN 'Oct'
    WHEN n_delays_per_month.month = 11 THEN 'Nov'
    WHEN n_delays_per_month.month = 12 THEN 'Dec'
  END AS month,
  n_flights_delayed,
  total_flights,
  ROUND((n_flights_delayed / total_flights) * 100, 1) AS percent_delayed
FROM
  n_delays_per_month
INNER JOIN
  total_flights_per_month
  ON n_delays_per_month.month = total_flights_per_month.month
ORDER BY
  percent_delayed DESC;
```

Analyzing the `flights` table revealed that Thursday, Friday, and Monday experience the highest number of flight delays, both cumulatively and proportionally.
```python
# Days of the week with most cancelled flights and percentage of severe delays

%%bigquery --project = ba775-b05

WITH week_day_delays AS
(
  SELECT
    day_of_week,
    SUM(CASE WHEN (departure_delay + arrival_delay) > 0 THEN 1 ELSE 0 END) AS n_flights_delayed,
    SUM(CASE WHEN (departure_delay + arrival_delay) > 60 THEN 1 ELSE 0 END) AS n_severe_delays,
    COUNT(*) AS total_flights
  FROM
    ba775-b05.airlines_kaggle.flights
  GROUP BY
    day_of_week
)

SELECT
  CASE
    WHEN day_of_week = 1 THEN 'Monday'
    WHEN day_of_week = 2 THEN 'Tuesday'
    WHEN day_of_week = 3 THEN 'Wednesday'
    WHEN day_of_week = 4 THEN 'Thursday'
    WHEN day_of_week = 5 THEN 'Friday'
    WHEN day_of_week = 6 THEN 'Saturday'
    WHEN day_of_week = 7 THEN 'Sunday'
  END AS day_of_week,
  n_flights_delayed,
  total_flights,
  ROUND((n_flights_delayed / total_flights) * 100, 1) AS percent_delayed,
  ROUND((n_severe_delays / total_flights) * 100, 1) AS percent_severe_delay
FROM
  week_day_delays
ORDER BY
  percent_delayed DESC;
```

We found that US Airways had the most with 6.3% cancellations and Atlantic Southeast came second with 3.4% cancellations. The rest of the airlines had cancellations in the 1-2% range. Additionally, we found that on October 17, 2015, US Airways merged with American Airlines ([Wikipedia](https://en.wikipedia.org/wiki/US_Airways)). Therefore, Atlantic Southeast is the only airline that should pay particular focus on reducing the number of cancellations for flights out of Boston, to be more in line with its competition. Therefore, at a system level, airlines face the issues of delayed flights more than cancellations.

```python
# Worst airlines to fly with (delays and cancellations)

%%bigquery --project = ba775-b05

SELECT
  Airline,
  total_flights,
  count_delays,
  ROUND((count_delays / total_flights) * 100, 2) AS perc_delays,
  cancellations,
  ROUND((cancellations / total_flights) * 100, 2) AS perc_cancelled
FROM
(
  SELECT
    A.Airline,
    COUNT(*) AS total_flights,
    SUM(IF(F.DEPARTURE_DELAY > 0, 1, 0)) AS count_delays,
    SUM(IF(F.CANCELLED > 0, 1, 0)) AS cancellations
  FROM
    ba775-b05.airlines_kaggle.flights F
    JOIN ba775-b05.airlines_kaggle.airlines A
    ON F.AIRLINE = A.IATA_Code
  WHERE
    ORIGIN_AIRPORT IN ('BOS', '10721')
  GROUP BY
    A.Airline
)
ORDER BY
  perc_delays DESC;
```

### **6.2 - Time Period Analysis**

#### **What are the months with highest number of delayed flights in Boston (cumulative and proportionally)?**

---

In our analysis of flights arriving and depparting from Boston, we identified that February, March, and June have the highest percentage of flight delays. February is the worst month with 51% of flights delayed. It is interesting to observe that when we analised the percentage of delays by month on a national level, the months with the most delays were June (1st), February (2nd), and July (3rd). The fact that February appears as the month with the most delays in Boston is probably a consequence of the bad weather that can be experienced during that month. Given that snowstorms can also happen in March, it is also a possibility that March flights experience higher delays in Boston when compared to national levels due to bad weather.

While insights are also valuable for resource allocation and management in airports and airlines for the Boston International Airport, the presence of snowstorms and bad weather in general during the earlier months of the year is a factor that might cause delays regardless of the level of preparedness of the airport and the local airlines.

**Note:** For this analysis we considered a flight to be delayed it the sum of its departure delay and arrival delay was larger than 1 (min).

```python
# Months with highest number of delayed flights for Boston flights

%%bigquery --project = ba775-b05

WITH n_delays_per_month AS(
  SELECT
    month,
    SUM(
      CASE
        WHEN (departure_delay + arrival_delay) > 0 THEN 1
        ELSE 0
        END
        ) AS n_flights_delayed,
  FROM ba775-b05.airlines_kaggle.boston_flights_vw
  GROUP BY month
),

total_flights_per_month AS(
  SELECT
    month,
    COUNT(*) AS total_flights
  FROM ba775-b05.airlines_kaggle.boston_flights_vw
  GROUP BY month
)

SELECT
  CASE
    WHEN n_delays_per_month.month = 1 THEN 'Jan' -- Changes the month values to a more readable format
    WHEN n_delays_per_month.month = 2 THEN 'Feb'
    WHEN n_delays_per_month.month = 3 THEN 'Mar'
    WHEN n_delays_per_month.month = 4 THEN 'Apr'
    WHEN n_delays_per_month.month = 5 THEN 'May'
    WHEN n_delays_per_month.month = 6 THEN 'Jun'
    WHEN n_delays_per_month.month = 7 THEN 'Jul'
    WHEN n_delays_per_month.month = 8 THEN 'Aug'
    WHEN n_delays_per_month.month = 9 THEN 'Sep'
    WHEN n_delays_per_month.month = 10 THEN 'Oct'
    WHEN n_delays_per_month.month = 11 THEN 'Nov'
    WHEN n_delays_per_month.month = 12 THEN 'Dec'
  END AS month,
  n_flights_delayed,
  total_flights,
  ROUND((n_flights_delayed/total_flights)*100,1) AS percent_delayed
FROM n_delays_per_month
INNER JOIN total_flights_per_month
ON n_delays_per_month.month = total_flights_per_month.month
ORDER BY percent_delayed DESC;
```

#### **What are the days of the week with highest number of delayed flights in Boston (cumulative and proportionally)?**

Analyzing only the flights arriving and departing from Boston, table revealed that Thursday, Friday, and Tuesday experience the highest number of flight delays, both cumulatively and proportionally. Specifically, Thursdays lead with the highest number of delays and the highest percentage of delayed flights (40.6%), followed closely by Fridays and Tuesdays.

On a national level, the days with the most delays are Thursday, Friday, and Monday. It is interesting that for Boston Tuesday is the 3rd day with the most delays instead of Monday, but the reason why is not clear. Additionally, Tuesday shows the highest percentage of severe delays at 13.9% of flights, which is significantly higher than the percentage of delays for Tuesdays on a national level at 10.6%.

From a business standpoint, the impact on airports and airlines is similar the earlier discussion about national weekday delays. However, a comparison of local Boston delay data with national trends reveals a clear need for more granular resource allocation strategies. This is because, as this analysis shows, the busiest weekdays and those with the most significant delays can differ between regions or individual airports. Therefore, airlines looking to optimize their resource allocation based on weekday or month (as seen in the previous question), might want to look into local patterns in addition to the national ones.

Note: For this analysis we considered a flight to be delayed it the sum of its departure delay and arrival delay was larger than 1 (min).

```python
# Days of the week with most cancelled flights for Boston flights

%%bigquery --project = ba775-b05

WITH week_day_delays AS
(
  SELECT
    day_of_week,
    SUM(CASE WHEN (departure_delay + arrival_delay) > 0 THEN 1 ELSE 0 END) AS n_flights_delayed,
    SUM(CASE WHEN (departure_delay + arrival_delay) > 60 THEN 1 ELSE 0 END) AS n_severe_delays,
    COUNT(*) AS total_flights
  FROM
    ba775-b05.airlines_kaggle.boston_flights_vw
  GROUP BY
    day_of_week
)

SELECT
  CASE
    WHEN day_of_week = 1 THEN 'Monday'
    WHEN day_of_week = 2 THEN 'Tuesday'
    WHEN day_of_week = 3 THEN 'Wednesday'
    WHEN day_of_week = 4 THEN 'Thursday'
    WHEN day_of_week = 5 THEN 'Friday'
    WHEN day_of_week = 6 THEN 'Saturday'
    WHEN day_of_week = 7 THEN 'Sunday'
  END AS day_of_week,
  n_flights_delayed,
  total_flights,
  ROUND((n_flights_delayed / total_flights) * 100, 1) AS percent_delayed,
  ROUND((n_severe_delays / total_flights) * 100, 1) AS percent_severe_delay
FROM
  week_day_delays
ORDER BY
  percent_delayed DESC;
```

#### **For out of Boston flights which airports recorded the worst delays by time of day?**

Delving into the delays by time of day for flights going out of Boston, we see that majority of `Evening flights` are delayed, followed by `Afternoon flights` and `Morning flights`.
```python
# Worst delays per time of day for flights out of Boston
# Checking for which airports contribute to the majority of these delays

%%bigquery --project = ba775-b05

select
c.departure_time_of_day,
c.origin_airport,
array_agg(c.dest_airport_code) destination_airport_codes,
array_agg(c.destination_airport) destination_airports,
c.tod_overall_dep_delay as tod_overall_avg_delay,
array_agg(c.tod_airport_dep_delay) destination_airport_avg_delay,
array_agg(c.tod_airport_delay_rank) destination_airport_rank
from
(
  select
  b.departure_time_of_day,
  b.origin_airport,
  b.destination_airport as dest_airport_code,
  ai.airport as destination_airport,
  ROUND(b.tod_overall_avg_delay,2) as tod_overall_dep_delay,
  ROUND(b.tod_airport_avg_delay,2) as tod_airport_dep_delay,
  b.tod_airport_delay_rank as tod_airport_delay_rank
  from
  (
    select a.*,
    RANK() OVER(partition by a.departure_time_of_day order by a.tod_airport_avg_delay desc) as tod_airport_delay_rank
    from
    (
      select distinct
      departure_time_of_day,
      origin_airport,
      destination_airport,
      AVG(departure_delay) OVER(partition by departure_time_of_day) as tod_overall_avg_delay,
      AVG(departure_delay) OVER(partition by departure_time_of_day, destination_airport) as tod_airport_avg_delay
      from airlines_kaggle.boston_departure_vw
      where departure_delay > 0
    ) a
  ) b
  join airlines_kaggle.airports ai
  on b.destination_airport = ai.IATA_CODE
  where b.tod_airport_delay_rank <=5
  order by tod_overall_dep_delay desc, b.tod_airport_avg_delay desc
) c
group by 1,2,5;
```

```python
# Creating view for departures from Boston


"""
%%bigquery --project = ba775-b05

CREATE VIEW IF NOT EXISTS airlines_kaggle.boston_departure_vw
AS
(
  select a.*,
  case when extract(time from scheduled_departure_dt) between '00:00:00' and '04:59:00'
            then 'Early Morning'
      when extract(time from scheduled_departure_dt) between '05:00:00' and '11:59:00'
            then 'Morning'
      when extract(time from scheduled_departure_dt) between '12:00:00' and '17:59:00'
            then 'Afternoon'
      when extract(time from scheduled_departure_dt) between '18:00:00' and '23:59:00'
            then 'Evening'
  end as departure_time_of_day
  from
  (
    select
    year,
    month,
    day,
    day_of_week,
    airline,
    flight_number,
    tail_number,
    origin_airport,
    destination_airport,
    scheduled_departure,
    parse_timestamp('%Y%m%d %k%M',
                    concat(year, LPAD(CAST(month as string), 2, '0'), day,
                    ' ', LPAD(cast(scheduled_departure as string), 4, '0'))) as scheduled_departure_dt,
    departure_time,
    departure_delay,
    case when departure_delay < 0 then 'Before time'
        when departure_delay = 0 then 'On time'
        when departure_delay > 0 then 'Delayed'
    end as status_of_departure,
    air_time,
    distance,
    diverted,
    cancelled
    from ba775-b05.airlines_kaggle.flights
    where origin_airport = 'BOS'
    and departure_delay is not null
  ) a
);
"""
```
